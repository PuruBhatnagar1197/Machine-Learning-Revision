```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"/content/hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv")
```

```python
df.head()
```

```python
X=df.iloc[:,3:5].values
```

```python
import scipy.cluster.hierarchy as shc
x=shc.dendrogram(shc.linkage(X,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cl=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
cl.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=cl.labels_,cmap='rainbow')
```
