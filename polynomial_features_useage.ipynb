```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel(r"random_test.xlsx")
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(df[['x','y']],df['class'])
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(df[['x','y']].values, df['class'].values.astype('int'), clf, legend=2)
```

```python
### how we are going to apply polynomialfeatures
from sklearn.preprocessing import PolynomialFeatures

pf_fe=PolynomialFeatures(degree=3,include_bias=False)
```

```python
x=pf_fe.fit_transform(df[['x','y']])
```

```python
clf.fit(x,df['class'])
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(clf,x,df['class'].values,scoring='accuracy',cv=10))
```
