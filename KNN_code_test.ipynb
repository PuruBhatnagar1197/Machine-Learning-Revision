```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(r"/content/data.csv")
```

```python
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df
```

```python
x_train, x_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_train_new=st.fit_transform(x_train)
x_test_new=st.transform(x_test)
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train_new,y_train)
```

```python
y_pred_1=knn.predict(x_test_new)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_1)
```

```python
score=[]
for i in range(1,16):
    knn1=KNeighborsClassifier(n_neighbors=i)
    knn1.fit(x_train_new,y_train)
    y_pred_sub=knn1.predict(x_test_new)
    score.append(accuracy_score(y_pred_sub,y_test))
```

```python
score
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,16),score)
```
