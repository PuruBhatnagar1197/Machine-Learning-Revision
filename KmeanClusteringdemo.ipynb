```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"/content/student_clustering.csv")
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss) ### cluster 4 is giving good result
```

```python
km1=KMeans(n_clusters=4)
```

```python
x=df.iloc[:,:].values
y_cluster=km1.fit_predict(x)
```

```python
plt.scatter(x[y_cluster==0,0],x[y_cluster==0,1],color='red')
plt.scatter(x[y_cluster==1,0],x[y_cluster==1,1],color='blue')
plt.scatter(x[y_cluster==2,0],x[y_cluster==2,1],color='orange')
plt.scatter(x[y_cluster==3,0],x[y_cluster==3,1],color='black')
```

### what if we want to do in 3D

```python
from sklearn.datasets import make_blobs
```

```python
centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]
x,y=make_blobs(n_samples=200,n_features=3,cluster_std=cluster_std,centers=centroids,random_state=1)
```

```python
wcss1=[]
for i in range(1,21):
    km2=KMeans(n_clusters=i)
    km2.fit(x)
    wcss1.append(km2.inertia_)
```

```python
plt.plot(range(1,21),wcss1)
```

```python
km2=KMeans(n_clusters=4)
y_cluster=km2.fit_predict(x)
```

```python
df = pd.DataFrame()
df['col1'] = x[:,0]
df['col2'] = x[:,1]
df['col3'] = x[:,2]
df['label'] = y_cluster
```

```python
import plotly.express as px
fig = px.scatter_3d(x=x[:,0], y=x[:,1], z=x[:,2])
fig.show()
```

```python
px.scatter_3d(df,x='col1',y='col2',z='col3',color='label')
```
